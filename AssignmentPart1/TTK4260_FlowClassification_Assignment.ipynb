## Context

Two-phase (gas–liquid) flow in pipelines exhibits distinct **flow regimes** — spatial arrangements of the gas and liquid phases — that critically affect pressure drop, heat transfer, and operational safety. Predicting the flow regime from measurable operating conditions is a classic engineering classification problem.

The dataset contains **2792 observations** from multiple experimental campaigns and mechanistic models, with **8 flow regime classes**:

| Abbreviation | Flow Regime | Samples |
|:---:|:---|:---:|
| SS | Stratified Smooth | 74 |
| SW | Stratified Wavy | 130 |
| AN | Annular | 362 |
| IN | Intermittent (Slug) | 1372 |
| BL | Bubbly | 454 |
| CH | Churn | 347 |
| DB | Dispersed Bubble | 35 |
| MS | Mist | 18 |

**Features** include superficial velocities ($v_{sl}$, $v_{sg}$), pipe geometry (ID, roughness, angle), fluid properties (densities, viscosities, surface tension), and thermodynamic conditions (pressure, temperature).
