## What Do We Actually Observe?

We observe $n$ data points with several explanatory variables:

| Observation | $x_1$ | $x_2$ | $x_3$ | Response $y$ |
|-------------|-------|-------|-------|-------------|
| 1 | $x_{11}$ | $x_{12}$ | $x_{13}$ | $y_1$ |
| 2 | $x_{21}$ | $x_{22}$ | $x_{23}$ | $y_2$ |
| ... | ... | ... | ... | ... |
| n | $x_{n1}$ | $x_{n2}$ | $x_{n3}$ | $y_n$ |

## Why This Was a Serious Problem (ca. 1800)

- Astronomers made repeated measurements
- Each measurement had error
- No single curve passed through all points

**The dilemma:** Which curve best represents the data?

*Exact fit is impossible — we must define what "best" means.*

````python
# Visualize the problem: noisy data with an underlying law
np.random.seed(42)
x_demo = np.linspace(0, 10, 30)
y_true = 2 + 0.5 * x_demo  # Underlying law
y_noisy = y_true + np.random.normal(0, 0.8, len(x_demo))  # Observed data with noise

fig = go.Figure()
fig.add_trace(go.Scatter(x=x_demo, y=y_noisy, mode='markers', name='Observed Data',
                         marker=dict(size=10, color='blue')))
fig.add_trace(go.Scatter(x=x_demo, y=y_true, mode='lines', name='Underlying Law',
                         line=dict(color='red', width=2)))
fig.update_layout(
    title='The Problem: Finding the Underlying Law from Noisy Data',
    xaxis_title='x',
    yaxis_title='y',
    font=dict(family="Times New Roman", size=14),
    showlegend=True
)
fig.show()
````

## Modeling Noisy Data: The Least Squares Idea

Assume a model with unknown parameters:

$$y_i \approx f(x_{i1}, x_{i2}, x_{i3}; \theta)$$

- $f$ is a known function (not necessarily linear)
- $\theta$ is a vector of unknown parameters
- More observations than parameters

### Least Squares Principle

Choose $\theta$ to minimize the total squared discrepancy:

$$\min_\theta \sum_{i=1}^{n} (y_i - f(x_{i1}, x_{i2}, x_{i3}; \theta))^2$$

# 2. Model Linear in Features and Parameters

A very important class of models you will want to be able to fit to data is of the form:

$$y = \theta_0 x_0 + \theta_1 x_1 + \ldots + \theta_k x_k + \varepsilon$$

In the equation the $\theta$'s are unknown parameters to be estimated.

### Important Distinction

**Linear in parameters ≠ Linear in features**

$y = \theta_0 + \theta_1 x + \theta_2 x^2$ is linear in $\theta$ but nonlinear in $x$!

````python
# Interactive visualization: Linear regression with adjustable parameters
def plot_linear_regression(theta0=2.0, theta1=0.5, noise_level=0.5, n_points=50):
    np.random.seed(42)
    x = np.linspace(0, 10, n_points)
    y_true = theta0 + theta1 * x
    y_noisy = y_true + np.random.normal(0, noise_level, n_points)
    
    # Create figure with subplots
    fig = make_subplots(rows=1, cols=1)
    
    # Add data points
    fig.add_trace(go.Scatter(x=x, y=y_noisy, mode='markers', name='Data Points',
                             marker=dict(size=8, color='blue', opacity=0.7)))
    
    # Add true line
    fig.add_trace(go.Scatter(x=x, y=y_true, mode='lines', name=f'True Line: y = {theta0:.1f} + {theta1:.1f}x',
                             line=dict(color='red', width=2)))
    
    # Add vertical offset lines (residuals)
    for i in range(0, len(x), 5):
        fig.add_trace(go.Scatter(x=[x[i], x[i]], y=[y_noisy[i], y_true[i]], 
                                 mode='lines', line=dict(color='gray', dash='dash', width=1),
                                 showlegend=False))
    
    fig.update_layout(
        title=f'Linear Model: ŷ = θ₀ + θ₁x (θ₀={theta0:.1f}, θ₁={theta1:.1f})',
        xaxis_title='x',
        yaxis_title='y',
        font=dict(family="Times New Roman", size=14),
        height=500
    )
    fig.show()

# Interactive widget
interact(
    plot_linear_regression,
    theta0=widgets.FloatSlider(value=2.0, min=-5, max=10, step=0.5, description='θ₀ (intercept)'),
    theta1=widgets.FloatSlider(value=0.5, min=-2, max=3, step=0.1, description='θ₁ (slope)'),
    noise_level=widgets.FloatSlider(value=0.5, min=0.1, max=2, step=0.1, description='Noise'),
    n_points=widgets.IntSlider(value=50, min=10, max=100, step=10, description='N Points')
);
````

## Geometric Interpretation

**Key Insight:** The fitted values $\hat{\mathbf{y}}$ are the **orthogonal projection** of $\mathbf{y}$ onto the column space of $\mathbf{X}$.

**Why orthogonal?**
- The residual vector $\mathbf{e} = \mathbf{y} - \hat{\mathbf{y}}$ is perpendicular to Col($\mathbf{X}$)
- This is the shortest distance from $\mathbf{y}$ to the subspace
- Mathematically: $\mathbf{X}^T \mathbf{e} = \mathbf{0}$

### Core Geometric Insight

**The residual vector $\mathbf{e} = \mathbf{y} - \hat{\mathbf{y}}$ is orthogonal to the column space of $\mathbf{X}$.**

This means: $\mathbf{X}^T \mathbf{e} = \mathbf{0}$

🎯 **Why this matters**: This orthogonality condition **uniquely determines** the OLS solution and guarantees it minimizes the sum of squared residuals.

````python
# Simple demonstration of orthogonality
def demonstrate_orthogonality():
    """Simple demonstration of the orthogonality condition"""
    # Simple example: 3 data points, 2 parameters
    X = np.array([[1, 1], [1, 2], [1, 3]])  # Design matrix
    y = np.array([2.1, 3.9, 6.1])          # Observed values
    
    # OLS solution
    theta = np.linalg.lstsq(X, y, rcond=None)[0]
    y_hat = X @ theta
    residual = y - y_hat
    
    print("📐 Orthogonality Demonstration:")
    print(f"   X = \n{X}")
    print(f"   y = {y}")
    print(f"   ŷ = {y_hat}")
    print(f"   residual = {residual}")
    print(f"\n🔍 Key Check - X'e (should be ≈ [0, 0]):")
    print(f"   X'e = {X.T @ residual}")
    print(f"\n✅ This confirms residual ⊥ Col(X)")

demonstrate_orthogonality()
````

### Mathematical Consequence

From the orthogonality condition $\mathbf{X}^T \mathbf{e} = \mathbf{0}$, we derive:

$$\mathbf{X}^T(\mathbf{y} - \mathbf{X}\boldsymbol{\hat{\theta}}) = \mathbf{0}$$

$$\mathbf{X}^T\mathbf{y} - \mathbf{X}^T\mathbf{X}\boldsymbol{\hat{\theta}} = \mathbf{0}$$

$$\boldsymbol{\hat{\theta}} = (\mathbf{X}^T\mathbf{X})^{-1}\mathbf{X}^T\mathbf{y}$$

**This is the normal equation - the closed-form OLS solution!**

## Cost Function

$$J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} (\hat{y}_i(\theta) - y_i)^2$$

### Terminology
This is also called the **Mean Squared Error (MSE)** when the factor is $\frac{1}{m}$ instead of $\frac{1}{2m}$. The $\frac{1}{2}$ is often included to simplify the derivative.

**In matrix notation:**

$$J(\boldsymbol{\theta}) = \frac{1}{2m} (\mathbf{y} - \mathbf{X}\boldsymbol{\theta})^T (\mathbf{y} - \mathbf{X}\boldsymbol{\theta}) = \frac{1}{2m} \|\mathbf{y} - \mathbf{X}\boldsymbol{\theta}\|^2$$

# 3. Optimization using Brute Force

Grid search evaluates the cost function at every point on a regular mesh in parameter space.

````python
# Generate sample data for demonstrations
np.random.seed(42)
n_samples = 50
X_data = np.random.uniform(0, 10, n_samples)
true_theta0, true_theta1 = 1.5, 2.0
y_data = true_theta0 + true_theta1 * X_data + np.random.normal(0, 1.5, n_samples)

# Design matrix with intercept
X_design = np.column_stack([np.ones(n_samples), X_data])

def cost_function(theta, X, y):
    """Compute MSE cost function"""
    m = len(y)
    predictions = X @ theta
    return (1 / (2 * m)) * np.sum((predictions - y) ** 2)

print(f"True parameters: θ₀ = {true_theta0}, θ₁ = {true_theta1}")
print(f"Number of samples: {n_samples}")
````

````python
# Interactive 3D visualization of cost function surface
def plot_cost_surface(grid_resolution=50):
    # Create parameter grid
    theta0_range = np.linspace(-1, 4, grid_resolution)
    theta1_range = np.linspace(0, 4, grid_resolution)
    THETA0, THETA1 = np.meshgrid(theta0_range, theta1_range)
    
    # Compute cost at each grid point
    COST = np.zeros_like(THETA0)
    for i in range(grid_resolution):
        for j in range(grid_resolution):
            theta = np.array([THETA0[i, j], THETA1[i, j]])
            COST[i, j] = cost_function(theta, X_design, y_data)
    
    # Find minimum
    min_idx = np.unravel_index(np.argmin(COST), COST.shape)
    min_theta0 = THETA0[min_idx]
    min_theta1 = THETA1[min_idx]
    min_cost = COST[min_idx]
    
    # Create 3D surface plot
    fig = go.Figure()
    
    fig.add_trace(go.Surface(
        x=THETA0, y=THETA1, z=COST,
        colorscale='Viridis',
        name='Cost Surface',
        opacity=0.9
    ))
    
    # Mark the minimum
    fig.add_trace(go.Scatter3d(
        x=[min_theta0], y=[min_theta1], z=[min_cost],
        mode='markers',
        marker=dict(size=10, color='red', symbol='diamond'),
        name=f'Minimum (θ₀={min_theta0:.2f}, θ₁={min_theta1:.2f})'
    ))
    
    fig.update_layout(
        title='Grid Search on Cost Function Surface',
        scene=dict(
            xaxis_title='θ₀',
            yaxis_title='θ₁',
            zaxis_title='MSE'
        ),
        font=dict(family="Times New Roman", size=14),
        height=600,
        width=800
    )
    fig.show()
    
    print(f"Grid search minimum: θ₀ = {min_theta0:.4f}, θ₁ = {min_theta1:.4f}")
    print(f"Minimum cost: {min_cost:.4f}")
    print(f"Total evaluations: {grid_resolution**2}")

interact(
    plot_cost_surface,
    grid_resolution=widgets.IntSlider(value=50, min=10, max=100, step=10, description='Grid Res.')
);
````

````python
# Contour plot view
def plot_cost_contour(grid_resolution=100):
    theta0_range = np.linspace(-1, 4, grid_resolution)
    theta1_range = np.linspace(0, 4, grid_resolution)
    THETA0, THETA1 = np.meshgrid(theta0_range, theta1_range)
    
    COST = np.zeros_like(THETA0)
    for i in range(grid_resolution):
        for j in range(grid_resolution):
            theta = np.array([THETA0[i, j], THETA1[i, j]])
            COST[i, j] = cost_function(theta, X_design, y_data)
    
    fig = go.Figure()
    
    fig.add_trace(go.Contour(
        x=theta0_range, y=theta1_range, z=COST,
        colorscale='Viridis',
        contours=dict(showlabels=True),
        name='Cost Contours'
    ))
    
    # Mark true parameters
    fig.add_trace(go.Scatter(
        x=[true_theta0], y=[true_theta1],
        mode='markers',
        marker=dict(size=15, color='red', symbol='star'),
        name=f'True (θ₀={true_theta0}, θ₁={true_theta1})'
    ))
    
    fig.update_layout(
        title='Cost Function Contour Plot',
        xaxis_title='θ₀',
        yaxis_title='θ₁',
        font=dict(family="Times New Roman", size=14),
        height=500,
        width=600
    )
    fig.show()

plot_cost_contour()
````

## Limitations of Grid Search

### Advantages:
- Simple to implement
- Guarantees finding global optimum (if grid is fine enough)
- Embarrassingly parallel
- Good for visualization

### Disadvantages:
- **Curse of dimensionality**: cost grows as $O(n^p)$ where $p$ is number of parameters
- Inefficient: evaluates many unnecessary points
- Resolution limited by grid spacing

### Example
With 100 grid points per dimension:
- 2 parameters: $100^2 = 10,000$ evaluations
- 5 parameters: $100^5 = 10^{10}$ evaluations
- 10 parameters: $100^{10} = 10^{20}$ evaluations (infeasible!)

# 4. Optimization using Gradient Descent

## Batch Gradient Descent

**Update Rule:**
$$\boldsymbol{\theta}_{t+1} = \boldsymbol{\theta}_t - \alpha \nabla_\theta J(\boldsymbol{\theta}_t)$$

**For least squares:**
$$\nabla_\theta J = \frac{1}{m} \mathbf{X}^T (\mathbf{X}\boldsymbol{\theta} - \mathbf{y})$$

**Key hyperparameter: Learning rate $\alpha$**
- Too small: slow convergence, many iterations needed
- Too large: overshooting, oscillation, divergence
- Just right: smooth convergence to optimum

````python
# Gradient descent implementation with visualization
def gradient_descent_linear(X, y, theta_init, learning_rate, n_iterations):
    """Batch gradient descent for linear regression"""
    m = len(y)
    theta = theta_init.copy()
    history = {'theta': [theta.copy()], 'cost': [cost_function(theta, X, y)]}
    
    for i in range(n_iterations):
        # Compute gradient
        gradient = (1/m) * X.T @ (X @ theta - y)
        # Update parameters
        theta = theta - learning_rate * gradient
        # Store history
        history['theta'].append(theta.copy())
        history['cost'].append(cost_function(theta, X, y))
    
    return theta, history
````

````python
# Interactive gradient descent visualization
def visualize_gradient_descent(theta0_init=0.0, theta1_init=0.0, learning_rate=0.01, n_iterations=50):
    theta_init = np.array([theta0_init, theta1_init])
    
    # Run gradient descent
    theta_final, history = gradient_descent_linear(X_design, y_data, theta_init, learning_rate, n_iterations)
    
    # Extract history
    theta_path = np.array(history['theta'])
    cost_path = np.array(history['cost'])
    
    # Create cost surface for background
    theta0_range = np.linspace(-1, 4, 100)
    theta1_range = np.linspace(0, 4, 100)
    THETA0, THETA1 = np.meshgrid(theta0_range, theta1_range)
    COST = np.zeros_like(THETA0)
    for i in range(100):
        for j in range(100):
            theta = np.array([THETA0[i, j], THETA1[i, j]])
            COST[i, j] = cost_function(theta, X_design, y_data)
    
    # Create subplots
    fig = make_subplots(rows=1, cols=2, 
                        subplot_titles=['Gradient Descent Path', 'Cost vs Iterations'],
                        specs=[[{"type": "contour"}, {"type": "scatter"}]])
    
    # Contour plot
    fig.add_trace(go.Contour(
        x=theta0_range, y=theta1_range, z=COST,
        colorscale='Viridis', showscale=False,
        contours=dict(showlabels=True)
    ), row=1, col=1)
    
    # Gradient descent path
    fig.add_trace(go.Scatter(
        x=theta_path[:, 0], y=theta_path[:, 1],
        mode='lines+markers',
        marker=dict(size=6, color='red'),
        line=dict(color='red', width=2),
        name='GD Path'
    ), row=1, col=1)
    
    # Mark start and end
    fig.add_trace(go.Scatter(
        x=[theta_path[0, 0]], y=[theta_path[0, 1]],
        mode='markers', marker=dict(size=15, color='green', symbol='star'),
        name='Start'
    ), row=1, col=1)
    
    fig.add_trace(go.Scatter(
        x=[theta_path[-1, 0]], y=[theta_path[-1, 1]],
        mode='markers', marker=dict(size=15, color='blue', symbol='diamond'),
        name='End'
    ), row=1, col=1)
    
    # Cost curve
    fig.add_trace(go.Scatter(
        x=list(range(len(cost_path))), y=cost_path,
        mode='lines+markers',
        marker=dict(size=4, color='blue'),
        line=dict(color='blue', width=2),
        name='Cost'
    ), row=1, col=2)
    
    fig.update_xaxes(title_text='θ₀', row=1, col=1)
    fig.update_yaxes(title_text='θ₁', row=1, col=1)
    fig.update_xaxes(title_text='Iteration', row=1, col=2)
    fig.update_yaxes(title_text='Cost (MSE)', row=1, col=2)
    
    fig.update_layout(
        title=f'Gradient Descent (α={learning_rate}, {n_iterations} iterations)',
        font=dict(family="Times New Roman", size=12),
        height=450, width=1000,
        showlegend=True
    )
    fig.show()
    
    print(f"Initial: θ₀ = {theta0_init:.4f}, θ₁ = {theta1_init:.4f}")
    print(f"Final: θ₀ = {theta_final[0]:.4f}, θ₁ = {theta_final[1]:.4f}")
    print(f"True: θ₀ = {true_theta0:.4f}, θ₁ = {true_theta1:.4f}")
    print(f"Final cost: {cost_path[-1]:.6f}")

interact(
    visualize_gradient_descent,
    theta0_init=widgets.FloatSlider(value=0.0, min=-1, max=4, step=0.5, description='Init θ₀'),
    theta1_init=widgets.FloatSlider(value=0.0, min=0, max=4, step=0.5, description='Init θ₁'),
    learning_rate=widgets.FloatLogSlider(value=0.01, base=10, min=-3, max=-0.5, step=0.1, description='α (lr)'),
    n_iterations=widgets.IntSlider(value=50, min=5, max=200, step=5, description='Iterations')
);
````

## Non-Convex Cost Functions

When the cost function is non-convex, gradient descent can get stuck in **local minima**.

## Effect of Feature Scaling on Optimization

**The Problem**: When features have vastly different scales, optimization becomes inefficient.

🎯 **Key Insight**: Features with larger scales dominate the cost function, while smaller-scale features get "ignored".

### Why Feature Scaling Matters:

1. **Unequal Learning Rates**: Different features need different step sizes
2. **Elongated Contours**: Cost function becomes "stretched" in some directions  
3. **Slow Convergence**: Gradient descent zigzags instead of taking direct path
4. **Numerical Instability**: Large gradients can cause overshooting  

````python
# Feature Scaling Effect Demonstration
def gradient_descent_linear(X, y, theta_init, learning_rate, iterations):
    """Simple gradient descent implementation"""
    m = len(y)
    theta = theta_init.copy()
    history = {'theta': [theta.copy()], 'cost': [cost_function(theta, X, y)]}
    
    for i in range(iterations):
        predictions = X @ theta
        errors = predictions - y
        gradient = (1/m) * X.T @ errors
        theta = theta - learning_rate * gradient
        
        history['theta'].append(theta.copy())
        history['cost'].append(cost_function(theta, X, y))
    
    return history

def demonstrate_scaling_effects(scale_factor=100):
    """
    Demonstrate how feature scaling affects gradient descent convergence
    """
    # Generate sample data
    np.random.seed(42)
    n = 50
    
    # Features with different scales
    x1 = np.random.uniform(0, 1, n)          # Small scale: [0, 1]
    x2 = np.random.uniform(0, scale_factor, n)  # Large scale: [0, scale_factor]
    
    # True relationship
    y = 2 + 3*x1 + 0.01*x2 + np.random.normal(0, 0.3, n)
    
    # Create design matrices
    X_unscaled = np.column_stack([np.ones(n), x1, x2])
    
    # Standardized features
    x1_std = (x1 - np.mean(x1)) / np.std(x1)
    x2_std = (x2 - np.mean(x2)) / np.std(x2)
    X_scaled = np.column_stack([np.ones(n), x1_std, x2_std])
    
    # Run gradient descent on both
    theta_init = np.array([1.0, 1.0, 0.005])
    history_unscaled = gradient_descent_linear(X_unscaled, y, theta_init, 0.01, 30)
    
    theta_init_scaled = np.array([1.0, 1.0, 0.5])  
    history_scaled = gradient_descent_linear(X_scaled, y, theta_init_scaled, 0.1, 30)
    
    # Plot comparison
    fig = make_subplots(rows=1, cols=2, 
                       subplot_titles=('Unscaled Features', 'Scaled Features'))
    
    # Convergence plots
    fig.add_trace(
        go.Scatter(x=list(range(len(history_unscaled['cost']))), 
                  y=history_unscaled['cost'],
                  mode='lines+markers', name='Unscaled',
                  line=dict(color='red', width=2)),
        row=1, col=1
    )
    
    fig.add_trace(
        go.Scatter(x=list(range(len(history_scaled['cost']))), 
                  y=history_scaled['cost'],
                  mode='lines+markers', name='Scaled',
                  line=dict(color='blue', width=2)),
        row=1, col=2
    )
    
    fig.update_xaxes(title_text='Iteration')
    fig.update_yaxes(title_text='Cost')
    fig.update_layout(
        height=400, 
        title_text=f'Feature Scaling Effect (Scale Ratio 1:{scale_factor})',
        font=dict(family="Arial", size=12)
    )
    fig.show()
    
    # Print results
    print(f"📊 Results (Scale Ratio 1:{scale_factor}):")
    print(f"   Unscaled final cost: {history_unscaled['cost'][-1]:.6f}")
    print(f"   Scaled final cost:   {history_scaled['cost'][-1]:.6f}")
    
    if history_scaled['cost'][-1] > 0:
        improvement = history_unscaled['cost'][-1] / history_scaled['cost'][-1]
        print(f"   ⚡ Improvement: {improvement:.1f}x better!")
    
    print(f"\n🎯 Key Insight:")
    print(f"   Feature x2 scale: 1 to {scale_factor}")
    print(f"   Unscaled GD: Struggles with mixed scales")
    print(f"   Scaled GD: Converges efficiently")

# Interactive widget
interact(
    demonstrate_scaling_effects,
    scale_factor=widgets.IntSlider(value=100, min=10, max=1000, step=50, 
                                  description='Scale Ratio')
);
````

## Feature Scaling Methods

**The Solution**: Scale features to similar ranges before optimization.

### 📈 **Benefits of Feature Scaling:**
- ✅ **Equal Contribution**: All features contribute equally to cost function
- ✅ **Faster Convergence**: Gradient descent takes more direct path to optimum
- ✅ **Larger Learning Rates**: Can use bigger steps without overshooting
- ✅ **Numerical Stability**: Prevents gradient explosion/vanishing
- ✅ **Better Performance**: Especially critical for neural networks & SVM

---

### **1. Standardization (Z-score Normalization)**
$$x' = \frac{x - \mu}{\sigma}$$

**Result**: Mean = 0, Standard Deviation = 1

**When to use**: 
- Data follows roughly normal distribution
- Want to preserve relative distances
- **Most common choice for gradient descent**

### **2. Min-Max Normalization** 
$$x' = \frac{x - x_{\min}}{x_{\max} - x_{\min}}$$

**Result**: Range = [0, 1]

**When to use**:
- Know the min/max bounds of your data
- Want all features in same range
- Sensitive to outliers!

### **3. Robust Scaling**
$$x' = \frac{x - \text{median}}{\text{IQR}}$$

**When to use**: Data has outliers (uses median & IQR instead of mean & std)

---

🎯 **Rule of Thumb**: Always standardize features when using gradient descent!

## Batch GD vs. Stochastic GD vs. Mini-batch GD

| Method | Batch Size | Updates/Epoch | Noise |
|--------|------------|---------------|-------|
| Batch GD | m (all data) | 1 | Low |
| Mini-batch GD | b (e.g., 32, 64) | m/b | Medium |
| Stochastic GD | 1 | m | High |

````python
# Compare different gradient descent variants
def compare_gd_variants(learning_rate=0.01, n_epochs=20, batch_size=10):
    np.random.seed(42)
    theta_init = np.array([0.0, 0.0])
    m = len(y_data)
    
    # Batch GD
    theta_batch = theta_init.copy()
    cost_batch = []
    for epoch in range(n_epochs):
        gradient = (1/m) * X_design.T @ (X_design @ theta_batch - y_data)
        theta_batch = theta_batch - learning_rate * gradient
        cost_batch.append(cost_function(theta_batch, X_design, y_data))
    
    # Stochastic GD
    theta_sgd = theta_init.copy()
    cost_sgd = []
    for epoch in range(n_epochs):
        indices = np.random.permutation(m)
        for i in indices:
            xi = X_design[i:i+1]
            yi = y_data[i:i+1]
            gradient = xi.T @ (xi @ theta_sgd - yi)
            theta_sgd = theta_sgd - learning_rate * gradient
        cost_sgd.append(cost_function(theta_sgd, X_design, y_data))
    
    # Mini-batch GD
    theta_mini = theta_init.copy()
    cost_mini = []
    for epoch in range(n_epochs):
        indices = np.random.permutation(m)
        for i in range(0, m, batch_size):
            batch_idx = indices[i:i+batch_size]
            xi = X_design[batch_idx]
            yi = y_data[batch_idx]
            gradient = (1/len(batch_idx)) * xi.T @ (xi @ theta_mini - yi)
            theta_mini = theta_mini - learning_rate * gradient
        cost_mini.append(cost_function(theta_mini, X_design, y_data))
    
    # Plot comparison
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=list(range(n_epochs)), y=cost_batch, mode='lines+markers',
                             name='Batch GD', line=dict(color='blue', width=2)))
    fig.add_trace(go.Scatter(x=list(range(n_epochs)), y=cost_sgd, mode='lines+markers',
                             name='Stochastic GD', line=dict(color='red', width=2)))
    fig.add_trace(go.Scatter(x=list(range(n_epochs)), y=cost_mini, mode='lines+markers',
                             name=f'Mini-batch GD (b={batch_size})', line=dict(color='green', width=2)))
    
    fig.update_layout(
        title='Comparison of Gradient Descent Variants',
        xaxis_title='Epoch',
        yaxis_title='Cost (MSE)',
        font=dict(family="Times New Roman", size=14),
        height=450
    )
    fig.show()

interact(
    compare_gd_variants,
    learning_rate=widgets.FloatLogSlider(value=0.01, base=10, min=-3, max=-0.5, step=0.1, description='α'),
    n_epochs=widgets.IntSlider(value=20, min=5, max=100, step=5, description='Epochs'),
    batch_size=widgets.IntSlider(value=10, min=1, max=25, step=1, description='Batch Size')
);
````

# 6. Feature Engineering

Feature engineering ($x \to \phi(x)$) is the process of transforming raw data into features that better represent the underlying problem to predictive models.

````python
# Polynomial feature engineering demonstration
def polynomial_feature_engineering(true_degree=3, fit_degree=3, noise=0.5, n_points=50):
    np.random.seed(42)
    x = np.linspace(0, 1, n_points)
    
    # Generate true polynomial data
    if true_degree == 1:
        y_true = 1 + 2*x
    elif true_degree == 2:
        y_true = 1 + 2*x - 3*x**2
    elif true_degree == 3:
        y_true = 1 + 2*x - 3*x**2 + 4*x**3
    else:
        y_true = 1 + 2*x - 3*x**2 + 4*x**3 - 2*x**4
    
    y_noisy = y_true + np.random.normal(0, noise, n_points)
    
    # Create polynomial features for fitting
    X_poly = np.column_stack([x**i for i in range(fit_degree + 1)])
    
    # Fit using normal equations
    theta = np.linalg.lstsq(X_poly, y_noisy, rcond=None)[0]
    y_pred = X_poly @ theta
    
    # Calculate R²
    ss_res = np.sum((y_noisy - y_pred)**2)
    ss_tot = np.sum((y_noisy - np.mean(y_noisy))**2)
    r2 = 1 - ss_res / ss_tot
    
    # Plot
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=x, y=y_noisy, mode='markers', name='Noisy Data',
                             marker=dict(size=8, color='blue', opacity=0.6)))
    fig.add_trace(go.Scatter(x=x, y=y_true, mode='lines', name='True Function',
                             line=dict(color='green', width=2, dash='dash')))
    fig.add_trace(go.Scatter(x=x, y=y_pred, mode='lines', name=f'Fitted (degree {fit_degree})',
                             line=dict(color='red', width=2)))
    
    fig.update_layout(
        title=f'Polynomial Fitting: True degree={true_degree}, Fit degree={fit_degree}, R²={r2:.4f}',
        xaxis_title='x',
        yaxis_title='y',
        font=dict(family="Times New Roman", size=14),
        height=450
    )
    fig.show()
    
    print(f"Fitted coefficients: {theta}")

interact(
    polynomial_feature_engineering,
    true_degree=widgets.IntSlider(value=3, min=1, max=4, step=1, description='True Degree'),
    fit_degree=widgets.IntSlider(value=3, min=1, max=10, step=1, description='Fit Degree'),
    noise=widgets.FloatSlider(value=0.5, min=0.1, max=2, step=0.1, description='Noise'),
    n_points=widgets.IntSlider(value=50, min=20, max=100, step=10, description='N Points')
);
````

# 7. Normal Equations

## Derivation

In matrix notation, the model can be written as:
$$\mathbf{y} = \mathbf{X}\boldsymbol{\theta}$$

The normal equations come from the orthogonality condition:
$$\mathbf{X}^T(\mathbf{y} - \hat{\mathbf{y}}) = \mathbf{0}$$

Substituting $\hat{\mathbf{y}} = \mathbf{X}\hat{\boldsymbol{\theta}}$:
$$\mathbf{X}^T(\mathbf{y} - \mathbf{X}\hat{\boldsymbol{\theta}}) = \mathbf{0}$$
$$\mathbf{X}^T\mathbf{X}\hat{\boldsymbol{\theta}} = \mathbf{X}^T\mathbf{y}$$

If $\mathbf{X}^T\mathbf{X}$ is invertible:
$$\hat{\boldsymbol{\theta}} = (\mathbf{X}^T\mathbf{X})^{-1}\mathbf{X}^T\mathbf{y}$$

````python
# Normal equations implementation and verification
def normal_equations_demo():
    # Compute using normal equations
    XtX = X_design.T @ X_design
    Xty = X_design.T @ y_data
    theta_normal = np.linalg.solve(XtX, Xty)
    
    # Compute using numpy lstsq
    theta_lstsq = np.linalg.lstsq(X_design, y_data, rcond=None)[0]
    
    # Compute predictions
    y_pred = X_design @ theta_normal
    residuals = y_data - y_pred
    
    print("Normal Equations Solution:")
    print(f"  θ₀ = {theta_normal[0]:.6f}")
    print(f"  θ₁ = {theta_normal[1]:.6f}")
    print(f"\nNumPy lstsq Solution:")
    print(f"  θ₀ = {theta_lstsq[0]:.6f}")
    print(f"  θ₁ = {theta_lstsq[1]:.6f}")
    print(f"\nTrue parameters: θ₀ = {true_theta0}, θ₁ = {true_theta1}")
    
    # Verify orthogonality: X'e = 0
    print(f"\nOrthogonality check (X'e): {X_design.T @ residuals}")
    
    # Calculate metrics
    mse = np.mean(residuals**2)
    r2 = 1 - np.sum(residuals**2) / np.sum((y_data - np.mean(y_data))**2)
    print(f"\nMSE: {mse:.6f}")
    print(f"R²: {r2:.6f}")
    
    # Plot
    x_line = np.linspace(0, 10, 100)
    y_line = theta_normal[0] + theta_normal[1] * x_line
    
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=X_data, y=y_data, mode='markers', name='Data',
                             marker=dict(size=8, color='blue')))
    fig.add_trace(go.Scatter(x=x_line, y=y_line, mode='lines', name='OLS Fit',
                             line=dict(color='red', width=2)))
    
    fig.update_layout(
        title='Normal Equations Solution',
        xaxis_title='x',
        yaxis_title='y',
        font=dict(family="Times New Roman", size=14),
        height=450
    )
    fig.show()

normal_equations_demo()
````

# 8. Non-Linear Least Squares

Least squares can also be used with **nonlinear models**:
$$y = \theta_0 (1 - e^{-\theta_1 x}) + \varepsilon$$

### Non-linear in Parameters
This model is **non-linear in parameters** because $\theta_1$ appears in the exponent. This means:
- No closed-form solution exists
- Must use iterative optimization (gradient descent, scipy, etc.)
- May have multiple local minima
- Solution depends on initialization

## Practical Example: Biochemical Oxygen Demand (BOD)

BOD is used as a measure of pollution produced by domestic and industrial wastes. The exponential model:
$$y = \theta_1 (1 - e^{-\theta_2 x})$$

- $\theta_1$: ultimate BOD
- $\theta_2$: overall rate constant
- $x$: incubation days
- $y$: measured BOD

````python
# BOD data from the lecture
bod_data = {
    'Observation': [1, 2, 3, 4, 5, 6],
    'Incubation_days': [1, 2, 3, 5, 7, 10],
    'BOD': [109, 149, 149, 191, 213, 224]
}
df_bod = pd.DataFrame(bod_data)

x_bod = np.array(df_bod['Incubation_days'])
y_bod = np.array(df_bod['BOD'])

print("BOD Dataset:")
print(df_bod.to_string(index=False))
````

````python
# Enhanced Non-linear least squares with detailed optimization visualization
def nonlinear_bod_model(x, theta1, theta2):
    """BOD model: y = θ₁(1 - e^(-θ₂x))"""
    return theta1 * (1 - np.exp(-theta2 * x))

def bod_cost_function(params, x, y):
    """Cost function for BOD model"""
    theta1, theta2 = params
    predictions = nonlinear_bod_model(x, theta1, theta2)
    return np.sum((y - predictions)**2)

def optimization_callback_factory():
    """Factory to create callback function for optimization tracking"""
    history = {'params': [], 'cost': []}
    
    def callback(params):
        history['params'].append(params.copy())
        cost = bod_cost_function(params, x_bod, y_bod)
        history['cost'].append(cost)
    
    return callback, history

def visualize_nonlinear_optimization(theta1_init=180, theta2_init=0.2, show_animation=True):
    """
    Comprehensive visualization of non-linear optimization process
    """
    # Create callback to track optimization
    callback_func, history = optimization_callback_factory()
    
    # Add initial point to history (this was missing!)
    initial_params = [theta1_init, theta2_init]
    history['params'].append(initial_params.copy())
    history['cost'].append(bod_cost_function(initial_params, x_bod, y_bod))
    
    # Create cost surface
    theta1_range = np.linspace(100, 300, 80)
    theta2_range = np.linspace(0.05, 0.8, 80)
    T1, T2 = np.meshgrid(theta1_range, theta2_range)
    
    COST = np.zeros_like(T1)
    for i in range(80):
        for j in range(80):
            COST[i, j] = bod_cost_function([T1[i, j], T2[i, j]], x_bod, y_bod)
    
    # Run optimization with tracking
    result = minimize(
        bod_cost_function, 
        [theta1_init, theta2_init], 
        args=(x_bod, y_bod), 
        method='L-BFGS-B',
        callback=callback_func,
        options={'disp': False}
    )
    
    # Extract optimization path
    param_history = np.array(history['params'])
    cost_history = np.array(history['cost'])
    theta1_opt, theta2_opt = result.x
    
    if show_animation:
        # Calculate dynamic ranges to include both cost grid and optimization path
        if len(param_history) > 0:
            # Get the bounds of the optimization path
            path_theta1_min, path_theta1_max = param_history[:, 0].min(), param_history[:, 0].max()
            path_theta2_min, path_theta2_max = param_history[:, 1].min(), param_history[:, 1].max()
            
            # Expand ranges to include both grid and path with generous padding
            padding_theta1 = max(20, (theta1_range.max() - theta1_range.min()) * 0.1)
            padding_theta2 = max(0.05, (theta2_range.max() - theta2_range.min()) * 0.1)
            
            theta1_plot_min = min(theta1_range.min(), path_theta1_min) - padding_theta1
            theta1_plot_max = max(theta1_range.max(), path_theta1_max) + padding_theta1
            theta2_plot_min = min(theta2_range.min(), path_theta2_min) - padding_theta2
            theta2_plot_max = max(theta2_range.max(), path_theta2_max) + padding_theta2
        else:
            # Fallback to grid ranges with padding
            padding_theta1 = (theta1_range.max() - theta1_range.min()) * 0.05
            padding_theta2 = (theta2_range.max() - theta2_range.min()) * 0.05
            theta1_plot_min, theta1_plot_max = theta1_range.min() - padding_theta1, theta1_range.max() + padding_theta1
            theta2_plot_min, theta2_plot_max = theta2_range.min() - padding_theta2, theta2_range.max() + padding_theta2
        
        # If the dynamic range extends beyond the original grid, recalculate contours for full coverage
        if (theta1_plot_min < theta1_range.min() or theta1_plot_max > theta1_range.max() or 
            theta2_plot_min < theta2_range.min() or theta2_plot_max > theta2_range.max()):
            
            # Extend grid to cover the full dynamic range
            theta1_extended = np.linspace(theta1_plot_min, theta1_plot_max, 100)
            theta2_extended = np.linspace(theta2_plot_min, theta2_plot_max, 100)
            T1_ext, T2_ext = np.meshgrid(theta1_extended, theta2_extended)
            
            COST_extended = np.zeros_like(T1_ext)
            for i in range(100):
                for j in range(100):
                    COST_extended[i, j] = bod_cost_function([T1_ext[i, j], T2_ext[i, j]], x_bod, y_bod)
            
            # Use extended grid for contour
            contour_x, contour_y, contour_z = theta1_extended, theta2_extended, COST_extended
        else:
            # Use original grid
            contour_x, contour_y, contour_z = theta1_range, theta2_range, COST
        
        # Create animated visualization
        fig = make_subplots(
            rows=2, cols=2,
            subplot_titles=[
                'Cost Landscape & Optimization Path',
                'Model Fitting Progress', 
                'Cost Evolution',
                'Parameter Evolution'
            ],
            specs=[[{"type": "contour"}, {"type": "scatter"}],
                   [{"type": "scatter"}, {"type": "scatter"}]]
        )
        
        # 1. Cost landscape with optimization path
        fig.add_trace(go.Contour(
            x=contour_x, y=contour_y, z=contour_z,
            colorscale='Reds', showscale=True,
            contours=dict(showlabels=True, labelfont=dict(size=8)),
            opacity=0.8,
            colorbar=dict(
                x=0.46, y=0.77, 
                len=0.35, thickness=12, 
                title="Cost (SSE)"
            )
        ), row=1, col=1)
        
        # Optimization path
        if len(param_history) > 0:
            fig.add_trace(go.Scatter(
                x=param_history[:, 0], y=param_history[:, 1],
                mode='lines+markers', 
                line=dict(color='blue', width=3),
                marker=dict(size=6, color='blue'),
                name='Optimization Path'
            ), row=1, col=1)
            
            # Mark start and end
            fig.add_trace(go.Scatter(
                x=[theta1_init], y=[theta2_init],
                mode='markers', 
                marker=dict(size=12, color='green', symbol='circle'),
                name='Start'
            ), row=1, col=1)
            
            fig.add_trace(go.Scatter(
                x=[theta1_opt], y=[theta2_opt],
                mode='markers', 
                marker=dict(size=15, color='red', symbol='star'),
                name=f'Optimum'
            ), row=1, col=1)
        
        # 2. Model fitting with intermediate curves
        x_smooth = np.linspace(0, 12, 100)
        
        # Data points
        fig.add_trace(go.Scatter(
            x=x_bod, y=y_bod, mode='markers', name='Data',
            marker=dict(size=12, color='black', symbol='circle')
        ), row=1, col=2)
        
        # Initial curve
        y_init = nonlinear_bod_model(x_smooth, theta1_init, theta2_init)
        fig.add_trace(go.Scatter(
            x=x_smooth, y=y_init, mode='lines',
            line=dict(color='green', width=2, dash='dash'),
            name='Initial Curve'
        ), row=1, col=2)
        
        # Final fitted curve
        y_final = nonlinear_bod_model(x_smooth, theta1_opt, theta2_opt)
        fig.add_trace(go.Scatter(
            x=x_smooth, y=y_final, mode='lines',
            line=dict(color='red', width=3),
            name='Final Curve'
        ), row=1, col=2)
        
        # Show some intermediate curves
        if len(param_history) > 5:
            # Select 3 evenly spaced intermediate points
            n_intermediate = min(3, len(param_history) // 3)
            indices = np.linspace(1, len(param_history)-2, n_intermediate, dtype=int)
            # Remove duplicates and sort
            indices = sorted(set(indices))
            
            for i, idx in enumerate(indices):
                theta1_int, theta2_int = param_history[idx]
                y_int = nonlinear_bod_model(x_smooth, theta1_int, theta2_int)
                fig.add_trace(go.Scatter(
                    x=x_smooth, y=y_int, mode='lines',
                    line=dict(color='orange', width=1, dash='dot'),
                    name='Intermediate Curves' if i == 0 else '',
                    showlegend=i == 0
                ), row=1, col=2)
        
        # 3. Cost evolution
        if len(cost_history) > 0:
            iterations = list(range(len(cost_history)))
            fig.add_trace(go.Scatter(
                x=iterations, y=cost_history,
                mode='lines+markers',
                line=dict(color='purple', width=2),
                marker=dict(size=4),
                name='Cost Evolution'
            ), row=2, col=1)
        
        # 4. Parameter evolution - simplified to avoid dual y-axis issues
        if len(param_history) > 0:
            iterations = list(range(len(param_history)))
            
            # Normalize parameters to [0,1] scale for better comparison
            theta1_norm = (param_history[:, 0] - param_history[:, 0].min()) / (param_history[:, 0].max() - param_history[:, 0].min()) if param_history[:, 0].max() != param_history[:, 0].min() else np.zeros_like(param_history[:, 0])
            theta2_norm = (param_history[:, 1] - param_history[:, 1].min()) / (param_history[:, 1].max() - param_history[:, 1].min()) if param_history[:, 1].max() != param_history[:, 1].min() else np.zeros_like(param_history[:, 1])
            
            fig.add_trace(go.Scatter(
                x=iterations, y=theta1_norm,
                mode='lines+markers',
                line=dict(color='blue', width=2),
                marker=dict(size=4),
                name='θ₁ (normalized)'
            ), row=2, col=2)
            
            fig.add_trace(go.Scatter(
                x=iterations, y=theta2_norm,
                mode='lines+markers',
                line=dict(color='red', width=2),
                marker=dict(size=4),
                name='θ₂ (normalized)'
            ), row=2, col=2)
        
        # Update axis labels and ranges
        fig.update_xaxes(title_text='θ₁', row=1, col=1, 
                        range=[theta1_plot_min, theta1_plot_max])
        fig.update_yaxes(title_text='θ₂', row=1, col=1,
                        range=[theta2_plot_min, theta2_plot_max])
        fig.update_xaxes(title_text='Incubation Days', row=1, col=2)
        fig.update_yaxes(title_text='BOD (mg/L)', row=1, col=2)
        fig.update_xaxes(title_text='Iteration', row=2, col=1)
        fig.update_yaxes(title_text='Cost (SSE)', row=2, col=1)
        fig.update_xaxes(title_text='Iteration', row=2, col=2)
        fig.update_yaxes(title_text='Normalized Parameter Value', row=2, col=2)
        
        fig.update_layout(
            title=f'Non-Linear Optimization Process: BOD Model',
            font=dict(family="Arial", size=11),
            height=800, width=1300,
            showlegend=True,
            # Ensure proper spacing and prevent cropping/overlap
            margin=dict(l=60, r=80, t=90, b=60)
        )
        
        fig.show()
    
    # Print detailed results
    print("🔍 Optimization Summary:")
    print("="*60)
    print(f"   Model: y = θ₁(1 - e^(-θ₂x))")
    print(f"   Initial parameters: θ₁ = {theta1_init:.1f}, θ₂ = {theta2_init:.3f}")
    print(f"   Optimal parameters: θ₁ = {theta1_opt:.2f}, θ₂ = {theta2_opt:.4f}")
    print(f"   Initial cost: {bod_cost_function([theta1_init, theta2_init], x_bod, y_bod):.2f}")
    print(f"   Final cost: {result.fun:.2f}")
    print(f"   Cost reduction: {((bod_cost_function([theta1_init, theta2_init], x_bod, y_bod) - result.fun) / bod_cost_function([theta1_init, theta2_init], x_bod, y_bod) * 100):.1f}%")
    print(f"   Iterations: {len(cost_history)}")
    print(f"   Convergence: {'✅ Success' if result.success else '❌ Failed'}")
    print("="*60)
    
    # Show model interpretation
    print(f"\n📊 Model Interpretation:")
    print(f"   θ₁ = {theta1_opt:.2f} represents the ultimate BOD value")
    print(f"   θ₂ = {theta2_opt:.4f} represents the reaction rate constant")
    print(f"   Half-saturation time: {np.log(2)/theta2_opt:.1f} days")

# Interactive demonstration
interact(
    visualize_nonlinear_optimization,
    theta1_init=widgets.FloatSlider(value=180, min=120, max=250, step=10, description='Init θ₁'),
    theta2_init=widgets.FloatSlider(value=0.2, min=0.05, max=0.6, step=0.05, description='Init θ₂'),
    show_animation=widgets.Checkbox(value=True, description='Show Full Analysis')
);
````

# 9. Weighted Least Squares

Weighted least squares (WLS) is a variation where residuals are weighted:

$$J(\boldsymbol{\theta}) = \frac{1}{2m} \sum_{i=1}^{m} w_i (y_i - \hat{y}_i(\boldsymbol{\theta}))^2$$

- $w_i$ is the weight assigned to the $i^{th}$ observation

### When to Use WLS?
- **Heteroscedasticity**: variance of errors varies with $x$
- Some observations are more reliable than others
- Dealing with aggregated data (different sample sizes)
- Prior knowledge about measurement precision

### Common Weight Choices
- $w_i = 1/\sigma_i^2$ (inverse variance)
- $w_i = 1/|y_i|$ (relative error)
- $w_i = n_i$ (sample size in group $i$)

### Matrix Form
$$J(\boldsymbol{\theta}) = (\mathbf{y} - \mathbf{X}\boldsymbol{\theta})^T \mathbf{W} (\mathbf{y} - \mathbf{X}\boldsymbol{\theta})$$

where $\mathbf{W} = \text{diag}(w_1, w_2, \ldots, w_m)$

### WLS Solution
$$\hat{\boldsymbol{\theta}}_{WLS} = (\mathbf{X}^T \mathbf{W} \mathbf{X})^{-1} \mathbf{X}^T \mathbf{W} \mathbf{y}$$

````python
# Weighted Least Squares demonstration
def wls_demonstration(heteroscedastic=True):
    np.random.seed(42)
    n = 100
    x = np.linspace(1, 10, n)
    
    # True model
    y_true = 2 + 1.5 * x
    
    if heteroscedastic:
        # Heteroscedastic errors: variance increases with x
        errors = np.random.normal(0, 0.5 * x, n)
        weights = 1 / (0.5 * x)**2  # Inverse variance weights
        title_suffix = " (Heteroscedastic)"
    else:
        # Homoscedastic errors
        errors = np.random.normal(0, 1, n)
        weights = np.ones(n)
        title_suffix = " (Homoscedastic)"
    
    y = y_true + errors
    X = np.column_stack([np.ones(n), x])
    
    # OLS solution
    theta_ols = np.linalg.lstsq(X, y, rcond=None)[0]
    y_pred_ols = X @ theta_ols
    
    # WLS solution
    W = np.diag(weights)
    theta_wls = np.linalg.solve(X.T @ W @ X, X.T @ W @ y)
    y_pred_wls = X @ theta_wls
    
    # Create subplots
    fig = make_subplots(rows=1, cols=2,
                        subplot_titles=['Data and Fits', 'Residuals Comparison'])
    
    # Data and fits
    fig.add_trace(go.Scatter(x=x, y=y, mode='markers', name='Data',
                             marker=dict(size=6, color='blue', opacity=0.5)), row=1, col=1)
    fig.add_trace(go.Scatter(x=x, y=y_pred_ols, mode='lines', name='OLS',
                             line=dict(color='red', width=2)), row=1, col=1)
    fig.add_trace(go.Scatter(x=x, y=y_pred_wls, mode='lines', name='WLS',
                             line=dict(color='green', width=2, dash='dash')), row=1, col=1)
    fig.add_trace(go.Scatter(x=x, y=y_true, mode='lines', name='True',
                             line=dict(color='black', width=1, dash='dot')), row=1, col=1)
    
    # Residuals
    res_ols = y - y_pred_ols
    res_wls = y - y_pred_wls
    fig.add_trace(go.Scatter(x=x, y=res_ols, mode='markers', name='OLS Residuals',
                             marker=dict(size=5, color='red', opacity=0.5)), row=1, col=2)
    fig.add_trace(go.Scatter(x=x, y=res_wls, mode='markers', name='WLS Residuals',
                             marker=dict(size=5, color='green', opacity=0.5)), row=1, col=2)
    fig.add_hline(y=0, line_dash="dash", line_color="black", row=1, col=2)
    
    fig.update_xaxes(title_text='x', row=1, col=1)
    fig.update_yaxes(title_text='y', row=1, col=1)
    fig.update_xaxes(title_text='x', row=1, col=2)
    fig.update_yaxes(title_text='Residuals', row=1, col=2)
    
    fig.update_layout(
        title=f'OLS vs Weighted Least Squares{title_suffix}',
        font=dict(family="Times New Roman", size=12),
        height=450, width=1000
    )
    fig.show()
    
    print(f"True parameters: θ₀ = 2.0, θ₁ = 1.5")
    print(f"OLS: θ₀ = {theta_ols[0]:.4f}, θ₁ = {theta_ols[1]:.4f}")
    print(f"WLS: θ₀ = {theta_wls[0]:.4f}, θ₁ = {theta_wls[1]:.4f}")

interact(
    wls_demonstration,
    heteroscedastic=widgets.Checkbox(value=True, description='Heteroscedastic')
);
````

# 10. Common Pitfalls

## 1. Outliers
- Squared errors heavily penalize outliers
- Single outlier can drastically change fit
- **Solution**: robust regression, remove outliers

## 2. Multicollinearity
- When features are highly correlated
- $\mathbf{X}^T\mathbf{X}$ becomes nearly singular
- **Solution**: remove features, regularization (Ridge)

## 3. Extrapolation
- Model may not be valid outside training range
- Predictions can be wildly wrong
- **Solution**: be cautious, use domain knowledge

## 4. Overfitting
- Too many features relative to samples
- Perfect fit on training, poor on test
- **Solution**: regularization, cross-validation

````python
# Demonstration of outlier effect
def outlier_demonstration(outlier_y=50, outlier_x=5):
    np.random.seed(42)
    x = np.linspace(0, 10, 30)
    y_true = 2 + 1.5 * x
    y = y_true + np.random.normal(0, 1, 30)
    
    # Add outlier
    x_with_outlier = np.append(x, outlier_x)
    y_with_outlier = np.append(y, outlier_y)
    
    # Fit without outlier
    X = np.column_stack([np.ones(30), x])
    theta_clean = np.linalg.lstsq(X, y, rcond=None)[0]
    
    # Fit with outlier
    X_outlier = np.column_stack([np.ones(31), x_with_outlier])
    theta_outlier = np.linalg.lstsq(X_outlier, y_with_outlier, rcond=None)[0]
    
    x_line = np.linspace(0, 10, 100)
    y_clean = theta_clean[0] + theta_clean[1] * x_line
    y_outlier = theta_outlier[0] + theta_outlier[1] * x_line
    
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=x, y=y, mode='markers', name='Clean Data',
                             marker=dict(size=8, color='blue')))
    fig.add_trace(go.Scatter(x=[outlier_x], y=[outlier_y], mode='markers', name='Outlier',
                             marker=dict(size=15, color='red', symbol='x')))
    fig.add_trace(go.Scatter(x=x_line, y=y_clean, mode='lines', name='Fit w/o outlier',
                             line=dict(color='green', width=2)))
    fig.add_trace(go.Scatter(x=x_line, y=y_outlier, mode='lines', name='Fit with outlier',
                             line=dict(color='red', width=2, dash='dash')))
    
    fig.update_layout(
        title='Effect of Outliers on OLS',
        xaxis_title='x',
        yaxis_title='y',
        font=dict(family="Times New Roman", size=14),
        height=450
    )
    fig.show()
    
    print(f"Without outlier: θ₀ = {theta_clean[0]:.4f}, θ₁ = {theta_clean[1]:.4f}")
    print(f"With outlier: θ₀ = {theta_outlier[0]:.4f}, θ₁ = {theta_outlier[1]:.4f}")

interact(
    outlier_demonstration,
    outlier_y=widgets.FloatSlider(value=50, min=0, max=100, step=5, description='Outlier Y'),
    outlier_x=widgets.FloatSlider(value=5, min=0, max=10, step=0.5, description='Outlier X')
);
````

## Evaluating the Fit: $R^2$ and Residual Analysis

### Coefficient of Determination ($R^2$)
$$R^2 = 1 - \frac{SS_{res}}{SS_{tot}} = 1 - \frac{\sum_i (y_i - \hat{y}_i)^2}{\sum_i (y_i - \bar{y})^2}$$

- $R^2 \in [0, 1]$ for models with intercept
- $R^2 = 1$: perfect fit
- $R^2 = 0$: model no better than mean

### Adjusted $R^2$
$$R^2_{adj} = 1 - \frac{(1 - R^2)(m - 1)}{m - p - 1}$$

Penalizes adding more features.

### Residual Analysis
- Plot residuals vs. fitted values
- Should show no pattern
- Check for: heteroscedasticity (funnel shape), non-linearity (curved pattern), outliers

> **Warning:** High $R^2$ does not guarantee a good model! Always check residuals and validate on held-out data.

````python
# Residual analysis visualization
def residual_analysis():
    # Use our fitted model
    theta = np.linalg.lstsq(X_design, y_data, rcond=None)[0]
    y_pred = X_design @ theta
    residuals = y_data - y_pred
    
    # Calculate metrics
    ss_res = np.sum(residuals**2)
    ss_tot = np.sum((y_data - np.mean(y_data))**2)
    r2 = 1 - ss_res / ss_tot
    n, p = X_design.shape
    r2_adj = 1 - (1 - r2) * (n - 1) / (n - p - 1)
    
    # Create subplots
    fig = make_subplots(rows=2, cols=2,
                        subplot_titles=['Fitted vs Actual', 'Residuals vs Fitted',
                                       'Residual Distribution', 'Q-Q Plot'])
    
    # Fitted vs Actual
    fig.add_trace(go.Scatter(x=y_data, y=y_pred, mode='markers', name='Data',
                             marker=dict(size=8, color='blue')), row=1, col=1)
    fig.add_trace(go.Scatter(x=[min(y_data), max(y_data)], y=[min(y_data), max(y_data)],
                             mode='lines', name='Perfect fit', line=dict(color='red', dash='dash')), 
                  row=1, col=1)
    
    # Residuals vs Fitted
    fig.add_trace(go.Scatter(x=y_pred, y=residuals, mode='markers', name='Residuals',
                             marker=dict(size=8, color='green')), row=1, col=2)
    fig.add_hline(y=0, line_dash="dash", line_color="black", row=1, col=2)
    
    # Residual histogram
    fig.add_trace(go.Histogram(x=residuals, name='Residuals', nbinsx=15,
                               marker_color='purple'), row=2, col=1)
    
    # Q-Q plot
    from scipy import stats
    sorted_residuals = np.sort(residuals)
    theoretical_quantiles = stats.norm.ppf(np.linspace(0.01, 0.99, len(residuals)))
    fig.add_trace(go.Scatter(x=theoretical_quantiles, y=sorted_residuals, mode='markers',
                             name='Q-Q', marker=dict(size=6, color='orange')), row=2, col=2)
    fig.add_trace(go.Scatter(x=[-3, 3], y=[-3*np.std(residuals), 3*np.std(residuals)],
                             mode='lines', line=dict(color='red', dash='dash')), row=2, col=2)
    
    fig.update_xaxes(title_text='Actual', row=1, col=1)
    fig.update_yaxes(title_text='Predicted', row=1, col=1)
    fig.update_xaxes(title_text='Predicted', row=1, col=2)
    fig.update_yaxes(title_text='Residuals', row=1, col=2)
    fig.update_xaxes(title_text='Residuals', row=2, col=1)
    fig.update_yaxes(title_text='Count', row=2, col=1)
    fig.update_xaxes(title_text='Theoretical Quantiles', row=2, col=2)
    fig.update_yaxes(title_text='Sample Quantiles', row=2, col=2)
    
    fig.update_layout(
        title=f'Residual Analysis (R² = {r2:.4f}, Adjusted R² = {r2_adj:.4f})',
        font=dict(family="Times New Roman", size=11),
        height=600, width=900,
        showlegend=False
    )
    fig.show()
    
    print(f"R² = {r2:.6f}")
    print(f"Adjusted R² = {r2_adj:.6f}")
    print(f"Mean of residuals: {np.mean(residuals):.6f}")
    print(f"Std of residuals: {np.std(residuals):.6f}")

residual_analysis()
````

# 11. Summary and Takeaways

## Key Equations

**Cost function:**
$$J = \frac{1}{2m} \|\mathbf{y} - \mathbf{X}\boldsymbol{\theta}\|^2$$

**OLS solution:**
$$\hat{\boldsymbol{\theta}} = (\mathbf{X}^T\mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$$

**GD update:**
$$\boldsymbol{\theta} := \boldsymbol{\theta} - \alpha \nabla J$$

## Key Concepts

- LS minimizes sum of squared residuals
- For linear-in-parameters: closed-form solution exists
- For non-linear: use iterative methods
- Under Gaussian noise: OLS = MLE
- Feature scaling improves optimization

## Method Selection Guide

| Situation | Recommended Method |
|-----------|-------------------|
| Small data, linear model | Normal equations |
| Need numerical stability | QR decomposition |
| Large data, convex | Mini-batch GD |
| Non-linear model | Scipy optimizer (L-BFGS-B) |
| Constraints needed | SLSQP |

````python
# Summary comparison of all methods
def summary_comparison():
    theta_init = np.array([0.0, 0.0])
    
    results = []
    
    # 1. Normal Equations
    import time
    start = time.time()
    theta_normal = np.linalg.lstsq(X_design, y_data, rcond=None)[0]
    time_normal = time.time() - start
    cost_normal = cost_function(theta_normal, X_design, y_data)
    results.append({'Method': 'Normal Equations', 'θ₀': theta_normal[0], 'θ₁': theta_normal[1], 
                   'Cost': cost_normal, 'Time (ms)': time_normal*1000})
    
    # 2. Gradient Descent
    start = time.time()
    history_gd = gradient_descent_linear(X_design, y_data, theta_init, 0.01, 1000)
    theta_gd = history_gd['theta'][-1]  # Get final parameters
    time_gd = time.time() - start
    cost_gd = cost_function(theta_gd, X_design, y_data)
    results.append({'Method': 'Gradient Descent', 'θ₀': theta_gd[0], 'θ₁': theta_gd[1], 
                   'Cost': cost_gd, 'Time (ms)': time_gd*1000})
    
    # 3. Scipy L-BFGS-B
    start = time.time()
    result_scipy = minimize(lambda t: cost_function(t, X_design, y_data), theta_init, method='L-BFGS-B')
    time_scipy = time.time() - start
    results.append({'Method': 'Scipy L-BFGS-B', 'θ₀': result_scipy.x[0], 'θ₁': result_scipy.x[1], 
                   'Cost': result_scipy.fun, 'Time (ms)': time_scipy*1000})
    
    # 4. QR Decomposition
    start = time.time()
    Q, R = np.linalg.qr(X_design)
    theta_qr = np.linalg.solve(R, Q.T @ y_data)
    time_qr = time.time() - start
    cost_qr = cost_function(theta_qr, X_design, y_data)
    results.append({'Method': 'QR Decomposition', 'θ₀': theta_qr[0], 'θ₁': theta_qr[1], 
                   'Cost': cost_qr, 'Time (ms)': time_qr*1000})
    
    df_results = pd.DataFrame(results)
    print("Comparison of Optimization Methods:")
    print("="*70)
    print(df_results.to_string(index=False, float_format='{:.6f}'.format))
    print("="*70)
    print(f"\nTrue parameters: θ₀ = {true_theta0}, θ₁ = {true_theta1}")

summary_comparison()
````
